```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
movie_name = []
year = []
time = []
rating = []
votes = []
gross = []
listIMDB = []
```

```python
url = 'https://www.imdb.com/list/ls561448893/?ref_=otl_1&page='
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
count = 1
while count<=30:
  urlFinal = url + str(count)
  count = count + 1
  response = requests.get(urlFinal)
  soup = BeautifulSoup(response.text, "html.parser")
  movie_data = soup.findAll('div',attrs={'class': 'lister-item mode-detail'})
  for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    year.append(year_of_release)
    time_duration = store.p.find('span', class_ = 'runtime').text
    time.append(time_duration)
    rating_of_movie = store.find('span', class_ = 'ipl-rating-star__rating').text
    rating.append(rating_of_movie)
    vote = store.find('span', attrs = {'name': 'nv'}).text
    votes.append(vote)
    # value = store.find_all('span', attrs = {'name': 'nv'})
    data = {
        "name": name,
        "Year of release":year_of_release,
        "Time duration":time_duration,
        "Rating of novie":rating_of_movie,
        "Vote":vote,
    }
    listIMDB.append(data)
```

```python
df = pd.DataFrame(listIMDB)
var = 1
for movie in listIMDB:
      print(str(var) + ' '+ movie['name'], '-', movie['Year of release'], '('+movie['Time duration'] +
          ') -', 'Rating of movie:', movie['Rating of novie'], movie['Vote'])
      var = var + 1
df.to_csv('imdb_top_3000_movies.csv',index=True)
```
